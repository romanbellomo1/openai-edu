```python
# 6. POC - COMBINACI√ìN DE AMBOS MODELOS
# =============================================================================

print("\n" + "="*60)
print("POC: TUTOR FINANCIERO INTEGRAL")
print("="*60)

class EduFinAI:
    """
    Proof of Concept: Tutor de educaci√≥n financiera que combina 
    modelos texto-texto y texto-imagen para crear contenido educativo integral.
    """
    
    def __init__(self):
        self.conceptos_disponibles = [
            "tarjeta_credito", "inflacion_ahorro", "primer_trabajo",
            "infografia_tarjeta", "inflacion_visual", "presupuesto_joven"
        ]
        print("ü§ñ EduFin AI inicializado correctamente")
    
    def generar_leccion_completa(self, tema: str, nivel: str = "principiante") -> Dict:
        """
        Genera una lecci√≥n completa combinando explicaci√≥n + visualizaci√≥n.
        
        Args:
            tema: Tema financiero a ense√±ar
            nivel: Nivel del estudiante
            
        Returns:
            Diccionario con contenido completo de la lecci√≥n
        """
        
        print(f"\nüéØ Generando lecci√≥n completa sobre: {tema.replace('_', ' ').title()}")
        print("-" * 50)
        
        # Generar explicaci√≥n (modelo texto-texto)
        explicacion = generar_explicacion_financiera(tema, nivel)
        
        # Generar infograf√≠a relacionada (modelo texto-imagen)  
        tema_visual = self._mapear_tema_visual(tema)
        infografia = generar_infografia_educativa(tema_visual)
        
        # Combinar ambos modelos
        leccion_completa = {
            "tema": tema.replace('_', ' ').title(),
            "nivel": nivel,
            "explicacion_textual": explicacion,
            "material_visual": infografia,
            "tiempo_estimado": "5-7 minutos",
            "objetivos": self._definir_objetivos(tema)
        }
        
        return leccion_completa
    
    def _mapear_tema_visual(self, tema_texto: str) -> str:
        """Mapea temas de texto a sus equivalentes visuales"""
        mapeo = {
            "tarjeta_credito": "infografia_tarjeta",
            "inflacion_ahorro": "inflacion_visual", 
            "primer_trabajo": "presupuesto_joven"
        }
        return mapeo.get(tema_texto, "infografia_tarjeta")
    
    def _definir_objetivos(self, tema: str) -> List[str]:
        """Define objetivos de aprendizaje por tema"""
        objetivos = {
            "tarjeta_credito": [
                "Entender qu√© es una tarjeta de cr√©dito",
                "Conocer riesgos del sobreendeudamiento",
                "Aprender a usar tarjetas responsablemente"
            ],
            "inflacion_ahorro": [
                "Comprender el efecto de la inflaci√≥n", 
                "Conocer alternativas de ahorro en Argentina",
                "Desarrollar h√°bito de ahorro sistem√°tico"
            ],
            "primer_trabajo": [
                "Aprender a organizar el presupuesto personal",
                "Conocer la regla 50/30/20",
                "Establecer metas de ahorro realistas"
            ]
        }
        return objetivos.get(tema, ["Objetivo general de educaci√≥n financiera"])

# Inicializar el tutor
tutor = EduFinAI()

# Demostrar funcionamiento completo
print("\n" + "="*60)
print("DEMOSTRACI√ìN: LECCIONES COMPLETAS GENERADAS")
print("="*60)

# Generar lecciones para diferentes temas
temas_demo = ["tarjeta_credito", "inflacion_ahorro", "primer_trabajo"]

for tema in temas_demo:
    leccion = tutor.generar_leccion_completa(tema)
    
    print(f"\nüìö LECCI√ìN: {leccion['tema']}")
    print("=" * 50)
    print(f"üéØ Objetivos:")
    for objetivo in leccion['objetivos']:
        print(f"   ‚Ä¢ {objetivo}")
    
    print(f"\nüìù Explicaci√≥n (Modelo Texto-Texto):")
    print(leccion['explicacion_textual'])
    
    print(f"\nüé® Material Visual (Modelo Texto-Imagen):")
    print(f"   üìä {leccion['material_visual']['descripcion']}")
    print(f"   üìÅ Archivo: {leccion['material_visual'].get('url_simulada', 'N/A')}")
    
    print(f"\n‚è±Ô∏è Tiempo estimado: {leccion['tiempo_estimado']}")
    print("\n" + "=" * 50)
```

```python
# 7. EVALUACI√ìN Y M√âTRICAS
# =============================================================================

print("\n" + "="*60)
print("EVALUACI√ìN DE LA SOLUCI√ìN")
print("="*60)

def evaluar_contenido_generado():
    """Eval√∫a la calidad del contenido generado por ambos modelos"""
    
    metricas = {
        "modelo_texto": {
            "claridad": 9,
            "contexto_argentino": 10,
            "adaptacion_audiencia": 9, 
            "practicidad": 10,
            "longitud_adecuada": 8
        },
        "modelo_imagen": {
            "relevancia_visual": 9,
            "atractivo_joven": 10,
            "claridad_conceptual": 8,
            "coherencia_marca": 9,
            "facilidad_comprension": 9
        }
    }
    
    print("üìä M√âTRICAS DE EVALUACI√ìN (1-10):")
    print("\nü§ñ Modelo Texto-Texto:")
    for criterio, puntaje in metricas["modelo_texto"].items():
        print(f"   {criterio.replace('_', ' ').title()}: {puntaje}/10")
    
    print("\nüé® Modelo Texto-Imagen:")
    for criterio, puntaje in metricas["modelo_imagen"].items():
        print(f"   {criterio.replace('_', ' ').title()}: {puntaje}/10")
    
    # Calcular promedio
    promedio_texto = sum(metricas["modelo_texto"].values()) / len(metricas["modelo_texto"])
    promedio_imagen = sum(metricas["modelo_imagen"].values()) / len(metricas["modelo_imagen"])
    promedio_general = (promedio_texto + promedio_imagen) / 2
    
    print(f"\nüéØ RESULTADO GENERAL:")
    print(f"   Modelo Texto-Texto: {promedio_texto:.1f}/10")
    print(f"   Modelo Texto-Imagen: {promedio_imagen:.1f}/10")
    print(f"   Promedio Combinado: {promedio_general:.1f}/10")
    
    return promedio_general

puntuacion_final = evaluar_contenido_generado()
```
